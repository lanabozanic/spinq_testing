```python
def processtomography2qubit(dataset):
    state_tomographies = dataset.copy()

    measurements = ["HH", "HV", "VH", "VV", # ZZ
                   "HD", "HA", "VD", "VA", # ZX
                   "HL", "HR", "VL", "VR", # ZY
                   "DH", "DV", "AH", "AV", # XZ
                   "DD", "DA", "AD", "AA", # XX
                   "DL", "DR", "AL", "AR", # XY
                   "LH", "LV", "RH", "RV", # YZ
                   "LD", "LA", "RD", "RA", # YX
                   "LL", "LR", "RL", "RR", # YY
                  ]
    
    # do quantum state tomography on every state in the data set, create a new dict to store these.
    for idx in state_tomographies:
        
        tomo_object = qst.QubitTomo(2)
        rho = tomo_object.qst_MLE(measurements, np.array((state_tomographies)[idx]))
        state_tomographies[idx] = rho.matrix
        
    choimatrixentries = []    
    # we need to construct |n><m|, according to the above formulas
    # construct E|n><m|
    for n in range(4):
        for m in range(4):
            if(n == m):
                entry = state_tomographies[str(n) + str(m)]
                choimatrixentries.append(entry)
            if(n < m):
                plusstate = state_tomographies["+" + str(n) + str(m)]
                imagstate = state_tomographies["i" + str(n) + str(m)]
                zerostate = state_tomographies[str(n) + str(n)]
                onestate = state_tomographies[str(m) + str(m)]
                
                entry = plusstate + 1j * imagstate - (1+1j)*(zerostate + onestate)/2
                choimatrixentries.append(entry)
            if(n > m):
                plusstate = state_tomographies["+" + str(m) + str(n)]
                imagstate = state_tomographies["i" + str(m) + str(n)]
                zerostate = state_tomographies[str(n) + str(n)]
                onestate = state_tomographies[str(m) + str(m)]
                
                entry = plusstate - 1j * imagstate - (1-1j)*(zerostate + onestate)/2
                choimatrixentries.append(entry)
    
    choi = np.block([[choimatrixentries[0], choimatrixentries[1], choimatrixentries[2],choimatrixentries[3]],
                     [choimatrixentries[4], choimatrixentries[5], choimatrixentries[6],choimatrixentries[7]],
                     [choimatrixentries[8], choimatrixentries[9], choimatrixentries[10],choimatrixentries[11]],
                     [choimatrixentries[12], choimatrixentries[13], choimatrixentries[14],choimatrixentries[15]]])
    
    return choi, choimatrixentries
```

```python
def getchoi(tomoresults):
    zero = tomoresults[0]
    one = tomoresults[1]
    plus = tomoresults[2]
    imag = tomoresults[3]
    
    choi00 = zero
    choi11 = one
    choi01 = plus + 1j * imag -(1+1j)*(zero + one)/2 
    choi10 = plus - 1j * imag -(1-1j)*(zero + one)/2 
    
    choi = np.block([[choi00, choi01], 
                     [choi10, choi11]])
    return choi

def processtomography1qubit(counts):
    measurements = ["H", "V", "D", "A", "L", "R"]
    tomo_results = []
    for count in counts:
        print(count)
        tomo_object = qst.QubitTomo(1)
        rho = tomo_object.qst_MLE(measurements, count)
        tomo_results.append(np.array(rho.matrix))
    return getchoi(tomo_results)
```

```python
def fidelitychi(th_choi, exp_choi):
    th = superoperator_transformations.superop2chi(th_choi)
    exp = superoperator_transformations.superop2chi(exp_choi)
    
    return np.trace(exp * th.conj().T)/(np.sqrt(np.trace(th.conj().T*th))*np.sqrt(np.trace(exp.conj().T*exp)))
```
