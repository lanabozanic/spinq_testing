```python
from qutip import *
import numpy as np
```

```python
b = qutip.Bloch()
```

```python
b.make_sphere()
```

```python
b.fig.clf()
vec = [0, 0, 1]
b.add_vectors(vec)
b.render()
```

```python
sqrtX = np.matrix([[1/np.sqrt(2),-1j/np.sqrt(2)], [-1j/np.sqrt(2),1/np.sqrt(2)]])
zero=np.matrix([[1], [0]])
```

```python
b.add_vectors(np.matrix.flatten(sqrtX * zero))
```

```python
b.render()
```
